```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)

y1 = 2*x
y2 = 3*x

plt.subplot(1,2,1) # 1 row 2 columns 1 index
plt.plot(x,y1,color='g',linestyle="solid",linewidth=5)

plt.subplot(1,2,2) # 1 row 2 columns 2 index
plt.plot(x,y2,color='r',linestyle="solid",linewidth=5)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)

y1 = 2*x
y2 = 3*x
y3=  4*x

plt.subplot(1,3,1) # 1 row , 3 column , 1 index
plt.plot(x,y1,color='g',linestyle="solid",linewidth=5)

plt.subplot(1,3,2) # 1 row , 3 column , 2 index
plt.plot(x,y2,color='r',linestyle="dotted",linewidth=5)

plt.subplot(1,3,3) # 1 row , 3 column , 3 index
plt.plot(x,y3,color='b',linestyle="dashed",linewidth=5)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)

y1 = 2*x
y2 = 3*x
y3=  4*x

plt.subplot(3,2,1)
plt.plot(x,y1,color='g',linestyle="solid",linewidth=5)

plt.subplot(3,2,2)
plt.plot(x,y2,color='r',linestyle="dotted",linewidth=5)

plt.subplot(3,1,3)
plt.plot(x,y3,color='b',linestyle="dashed",linewidth=5)

plt.show()
```

```python
from matplotlib import pyplot as plt

rollno=[1,2,3,4,5,6,7,8,9]
hindi=[11,12,19,14,13,16,13,10,19]
english=[15,10,18,10,13,14,15,19,12]

plt.subplot(1,2,1)
plt.scatter(hindi,rollno, marker="*",color="red",s=500)
plt.xlabel("Hindi Marks")
plt.ylabel("Rollno")
plt.subplot(1,2,2)
plt.scatter(english,rollno, marker="s",color="blue",s=200)
plt.xlabel("English Marks")
plt.ylabel("Rollno")

plt.grid(True)
plt.show()
```
